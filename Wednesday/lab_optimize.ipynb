```python
import pickle
import random
import numpy as np
import itertools
from sklearn.tree import DecisionTreeRegressor


with open("auction.pickle", "rb") as auction_file:
    AUCTIONS = pickle.load(auction_file)
print(AUCTIONS[0])

ITEM_TYPES = set()
for round in AUCTIONS:
    for item in round[0]:
        ITEM_TYPES.add(item)
print(ITEM_TYPES)
        
        
DATASETS = dict()
TARGETS = dict()
TREES = dict()

def emptyData():
    global DATASETS, TARGETS, TREES
    DATASETS = dict()
    TARGETS = dict()
    TREES = dict()
    for t in ITEM_TYPES:
        DATASETS[t] = []
        TARGETS[t] = []
        TREES[t] = DecisionTreeRegressor(min_samples_leaf=10)

def createRow(item, sold, remain, bids):
    newrow = []

    newrow = newrow + [sum(bids)]
    newrow = newrow + [len(sold)]

    for t in ITEM_TYPES:
        newrow = newrow + [sold.count(t)]
        newrow = newrow + [remain.count(t)]

    sumbids = dict()
    for t in ITEM_TYPES:
        sumbids[t] = 0.0
    for i in range(len(sold)):
        sumbids[sold[i]] = sumbids[sold[i]] + bids[i]
    for t in ITEM_TYPES:
        newrow = newrow + [sumbids[t]]

    #########################################
    ### TODO: also add new features here! ###
    #########################################

    return newrow

def addRow(bid, item, sold, remain, bids):
    global DATASETS, TARGETS, TREES
    TARGETS[item] = TARGETS[item] + [bid]
    DATASETS[item] = DATASETS[item] + [createRow(item, sold, remain, bids)]

def addRows(items, bids):
    remain = list(items)
    done_bids = []
    sold = []
    while len(remain) != 0:
        item = remain.pop(0)
        bid = bids.pop(0)

        addRow(bid, item, sold, remain, done_bids)

        sold = sold + [item]
        done_bids = done_bids + [bid]        
                
emptyData()
for round in AUCTIONS:
    addRows(round[0],round[1])
```

```python
def learnTrees():
    global DATASETS, TARGETS, TREES
    for t in ITEM_TYPES:
        TREES[t].fit(DATASETS[t],TARGETS[t])

def evaluateOrder(order):
    values = []
    sold = []
    remain = list(order)
    while remain != []:
        item = remain.pop(0)
        row = createRow(item, sold, remain, values)
        values = values + [TREES[item].predict(np.array(row).reshape(1, -1))[0]]
        sold = sold + [item]
        #print(row)
    return values

learnTrees()

itemtype = ITEM_TYPES
items_per_type = 3
nrItems = dict(zip(itemtype,[items_per_type]*len(itemtype)))
index = list(range(items_per_type*len(itemtype)))

test_order = list()
for item in nrItems:
    for i in range(nrItems[item]):
        test_order.append(item)
        
print(test_order)
for i in range(10):
    random.shuffle(test_order)
    print(test_order)
    result = evaluateOrder(test_order)
    print(str(result) + " " + str(sum(result)))
```

```python
import gurobipy as gp
from gurobipy import GRB

M = 1000000

def featuremap():
    features = []
    values = []
    features.append("sumv")
    values.append(M)
    features.append("index")
    values.append(len(index))

    for t in ITEM_TYPES:
        features.append("sold_" + t)
        values.append(nrItems[t])
        features.append("remain_" + t)
        values.append(nrItems[t])

    for t in ITEM_TYPES:
        features.append("sum_" + t)
        values.append(M)

    return features, values

def find_path(tree, i):
    if i == 0:
        return []
    for j in range(len(tree.children_left)):
        if tree.children_left[j] == i:
            return find_path(tree,j) + [j]
    for j in range(len(tree.children_right)):
        if tree.children_right[j] == i:
            return find_path(tree,j) + [-j]

def find_children(tree, i):
    if tree.children_left[i] == -1:
        return [i]
    return find_children(tree, tree.children_left[i]) + find_children(tree, tree.children_right[i])
      
def left_children(t, i):
    tree = TREES[t].tree_
    if tree.children_left[i] == -1:
        return []
    return find_children(tree, tree.children_left[i])

def right_children(t, i):
    tree = TREES[t].tree_
    if tree.children_right[i] == -1:
        return []
    return find_children(tree, tree.children_right[i])

def value(l, t):
    tree = TREES[t].tree_
    return tree.value[l][0][0]

def feature(n, t):
    tree = TREES[t].tree_
    return tree.feature[n]

def threshold(n, t):
    tree = TREES[t].tree_
    return tree.threshold[n]


def lpdtree():
    model = gp.Model("Decision_Tree_Optimization")
    
    nodes = dict()
    for t in itemtype:
        nodes[t] = []
    leafs = dict()
    for t in itemtype:
        leafs[t] = []
        
    for t in itemtype:
        tree = TREES[t].tree_
        for n in range(len(tree.value)):
            if tree.children_left[n] != -1:
                nodes[t].append(n)
            else:
                leafs[t].append(n)
    

    # an itemtype t item is auctioned at index i
    x = model.addVars(index,itemtype, name="x",vtype=GRB.BINARY)
    # node n evaluates to true at index i
    y = dict()
    for t in itemtype:
        y[t] = model.addVars(index,nodes[t], name="y"+str(t),vtype=GRB.BINARY)
    # leaf l evaluates to true at index i
    z = dict()
    for t in itemtype:
        z[t] = model.addVars(index,leafs[t], name="z"+str(t),vtype=GRB.BINARY)
    # the trees predict value v at index i
    v = model.addVars(index, itemtype, name="v",vtype=GRB.CONTINUOUS)
    
    # the value of feature f at index i
    feat = dict()
    features, maxvalues = featuremap()
    for f in range(len(features)):
        feat[features[f]] = model.addVars(index,name=features[f],vtype=GRB.CONTINUOUS,lb=0,ub=maxvalues[f])

    # Create the constraints
    model.modelSense = GRB.MAXIMIZE

    model.setObjective(gp.quicksum(v[i,t] for i in index for t in itemtype))

    #########################################
    ### TODO: place your constraints here ###
    #########################################

   #This is the tree formulation
    for i in index:
        for t in itemtype:
            model.addConstr(gp.quicksum(z[t][i,l] for l in leafs[t]) == x[i,t])
            model.addConstr(v[i,t] == gp.quicksum(value(l,t) * z[t][i,l] for l in leafs[t]))

    for t in itemtype:
        for i in index:
            for n in nodes[t]:
                model.addConstr(gp.quicksum(z[t][i,l] for l in left_children(t,n)) <= 1 - y[t][i,n])
                model.addConstr(gp.quicksum(z[t][i,l] for l in right_children(t,n)) <= y[t][i,n])
                           
    for t in itemtype:
        for i in index:
            for n in nodes[t]:
                f = feature(n,t)
                model.addConstr(y[t][i,n]*threshold(n,t) <= feat[features[f]][i])
                model.addConstr(threshold(n,t) + y[t][i,n]*maxvalues[f] >= feat[features[f]][i])

    model.write("LPtree.lp")
    model.optimize()
    print("Status: ", model.Status)

    itemorder = []
    values = []
    if model.Status == GRB.OPTIMAL:
        for i in index:
            for t in itemtype:
                if x[i,t].X == 1:
                    itemorder.append(t)
        for i in index:
            val = 0
            for t in itemtype:
                val += v[i,t].X
            values.append(val)
        print("Objective = " + str(model.ObjVal))
        print(itemorder)
        print("expected: " + str(values))
        #for i in index:
        #    print([feat[features[f]][i].X for f in range(len(features))])
        actual_result = evaluateOrder(itemorder)
        print("actual: " + str(actual_result))
        print("actual Objective: " + str(sum(actual_result)))
        
        item_file = open("items.pickle", "wb")
        pickle.dump(itemorder, item_file)     
              

lpdtree()
```
